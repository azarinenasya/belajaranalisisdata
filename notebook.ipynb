## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

**Contoh pertanyaan bisnis yang memenuhi seluruh elemen SMART**

***"Faktor apa saja yang memengaruhi penurunan conversion rate pada pengguna aplikasi Android di wilayah Jabodetabek sebesar 5% selama periode Flash Sale Maret 2026?"***

Keterangan:

- **Specific**: Fokus pada "penurunan conversion rate" untuk "aplikasi Android" di "Jabodetabek". Bukan sekadar penjualan turun.
- **Measurable**: Ada angka konkret yang ingin dianalisis, yaitu penurunan sebesar "5%".
- **Action-Oriented**: Dengan mengetahui faktor penyebabnya misalnya bug pada versi Android tertentu atau kendala logistik di Jabodetabek, tim bisa langsung melakukan perbaikan teknis atau operasional.
- **Relevant**: Penurunan konversi saat Flash Sale adalah masalah kritis bagi bisnis retail/e-commerce.
- **Time-bound**: Dibatasi pada periode spesifik "Flash Sale Maret 2026".

```python
# Mengidentifikasi dan mengklasifikasikan setiap stasiun sebagai 'pusat kota' atau 'pinggiran'
if 'station' in combined_df.columns and 'location_type' not in combined_df.columns:
    suburban_stations = ['Changping', 'Dingling', 'Huairou', 'Shunyi', 'Wanliu']
    city_center_stations = ['Aotizhongxin', 'Dongsi', 'Guanyuan', 'Gucheng', 'Nongzhanguan', 'Tiantan', 'Wanshouxigong']

    def classify_station_location(station_name):
        if station_name in suburban_stations:
            return 'Suburban'
        elif station_name in city_center_stations:
            return 'City Center'
        else:
            return 'Unknown'

    combined_df['location_type'] = combined_df['station'].apply(classify_station_location)
    print("  -> Kolom 'location_type' untuk stasiun berhasil dibuat.")
else:
    print("  -> Kolom 'location_type' sudah ada atau kolom 'station' tidak ditemukan.")
```

#### Fixing lost value

```python
# Menangani nilai yang hilang (missing values)
print("\n--- Menangani Nilai yang Hilang ---")
print("Jumlah nilai yang hilang sebelum imputasi:")
initial_missing = combined_df.isnull().sum()[combined_df.isnull().sum() > 0]
if not initial_missing.empty:
    print(initial_missing)
else:
    print("Tidak ada nilai yang hilang sebelum imputasi.")

numerical_cols = combined_df.select_dtypes(include=np.number).columns.tolist()
categorical_cols = combined_df.select_dtypes(include='object').columns.tolist()

# Hapus kolom 'No' dari daftar numerik untuk imputasi jika ada
if 'No' in numerical_cols:
    numerical_cols.remove('No')

print("\n  -> Melakukan imputasi kolom numerik dengan median...")
for col in numerical_cols:
    if combined_df[col].isnull().any():
        median_val = combined_df[col].median()
        combined_df[col].fillna(median_val, inplace=True)
        print(f"    Kolom '{col}' diimputasi dengan median: {median_val}")

print("\n  -> Melakukan imputasi kolom kategorikal dengan modus...")
for col in categorical_cols:
    if combined_df[col].isnull().any():
        # Untuk 'wd' yang memiliki missing values, kita isi dengan 'Unknown' agar tidak bias pada modus
        if col == 'wd':
            combined_df[col].fillna('Unknown', inplace=True)
            print(f"    Kolom '{col}' diimputasi dengan 'Unknown'.")
        else:
            mode_val = combined_df[col].mode()[0]
            combined_df[col].fillna(mode_val, inplace=True)
            print(f"    Kolom '{col}' diimputasi dengan modus: {mode_val}")
```

```python
print("\nJumlah nilai yang hilang setelah imputasi:")
missing_after_imputation = combined_df.isnull().sum()[combined_df.isnull().sum() > 0]
if missing_after_imputation.empty:
    print("Tidak ada nilai yang hilang setelah imputasi.")
else:
    print(missing_after_imputation)

print("\n--- Ringkasan DataFrame setelah Preprocessing dan Cleaning ---")
combined_df.info()
print("\n5 baris pertama dari combined_df setelah semua preprocessing:")
print(combined_df.head())
```

```python
# 1. Filter data untuk lokasi pusat kota
city_center_df = combined_df[combined_df['location_type'] == 'City Center'].copy()

# 2. Definisikan kategori hari: hari kerja dan akhir pekan
city_center_df['day_type'] = city_center_df['is_weekday'].apply(lambda x: 'Weekday' if x else 'Weekend')

# 3. Definisikan jam sibuk dan non-sibuk
rush_hour_morning = (city_center_df['hour'] >= 7) & (city_center_df['hour'] < 10)
rush_hour_evening = (city_center_df['hour'] >= 16) & (city_center_df['hour'] < 19)
city_center_df['is_rush_hour'] = rush_hour_morning | rush_hour_evening

# 4. Analisis rata-rata CO berdasarkan hari kerja/akhir pekan untuk semua jam
print("\n--- Rata-rata CO di Pusat Kota (Semua Jam) ---")
co_all_hours = city_center_df.groupby('day_type')['CO'].mean().reset_index()
print(co_all_hours)

# 5. Analisis rata-rata CO berdasarkan hari kerja/akhir pekan untuk jam sibuk
print("\n--- Rata-rata CO di Pusat Kota (Jam Sibuk) ---")
co_rush_hours = city_center_df[city_center_df['is_rush_hour']].groupby('day_type')['CO'].mean().reset_index()
print(co_rush_hours)

# 6. Analisis rata-rata CO berdasarkan hari kerja/akhir pekan untuk jam non-sibuk
print("\n--- Rata-rata CO di Pusat Kota (Jam Non-Sibuk) ---")
co_non_rush_hours = city_center_df[~city_center_df['is_rush_hour']].groupby('day_type')['CO'].mean().reset_index()
print(co_non_rush_hours)
```

```python
# Filter data untuk hari kerja di musim dingin
winter_weekday_df = combined_df[(combined_df['is_weekday'] == True) & (combined_df['season'] == 'Winter')].copy()

# Total jam dan hari unik yang diamati selama hari kerja di musim dingin
total_winter_weekday_hours = len(winter_weekday_df)
total_winter_weekday_days = winter_weekday_df['datetime'].dt.date.nunique()

print(f"Data difilter untuk hari kerja di musim dingin. Jumlah data: {len(winter_weekday_df)} baris.")
print(f"Total jam di hari kerja musim dingin: {total_winter_weekday_hours}")
print(f"Total hari unik di hari kerja musim dingin: {total_winter_weekday_days}")

# Mengidentifikasi data di mana konsentrasi NO2 > 80 µg/m³ DAN kecepatan angin (WSPM) < 1.5 m/s
condition_met_df = winter_weekday_df[
    (winter_weekday_df['NO2'] > 80) &
    (winter_weekday_df['WSPM'] < 1.5)
].copy()

print(f"Data memenuhi kondisi NO2 > 80 µg/m³ dan WSPM < 1.5 m/s: {len(condition_met_df)} baris.")

# Menghitung total jam (jumlah baris) yang memenuhi kondisi ini
total_hours_condition_met = len(condition_met_df)
print(f"Total jam yang memenuhi kondisi: {total_hours_condition_met}")

# Menghitung total hari unik yang memenuhi kondisi ini
total_days_condition_met = condition_met_df['datetime'].dt.date.nunique()
print(f"Total hari unik yang memenuhi kondisi: {total_days_condition_met}")

# Menghitung persentase jam dan hari tersebut terhadap total jam dan hari yang diamati selama hari kerja di musim dingin
percentage_hours = (total_hours_condition_met / total_winter_weekday_hours) * 100 if total_winter_weekday_hours > 0 else 0
percentage_days = (total_days_condition_met / total_winter_weekday_days) * 100 if total_winter_weekday_days > 0 else 0

print(f"\nPersentase waktu (dalam jam) kondisi terpenuhi: {percentage_hours:.2f}%")
print(f"Persentase total hari kondisi terpenuhi: {percentage_days:.2f}%")
```
